```python
# Sample data (replace with your data)
square_footage = [1000, 1500, 2000, 2500, 3000, 3500]
prices = [150000, 200000, 250000, 300000, 350000, 400000]

# Calculate the mean of square footage and prices
mean_sqft = sum(square_footage) / len(square_footage)
mean_price = sum(prices) / len(prices)

# Calculate the coefficients (slope and intercept)
numerator = sum([(sqft - mean_sqft) * (price - mean_price) for sqft, price in zip(square_footage, prices)])
denominator = sum([(sqft - mean_sqft) ** 2 for sqft in square_footage])
slope = numerator / denominator
intercept = mean_price - (slope * mean_sqft)

# Print the coefficients
print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")

# Make predictions
predicted_prices = [slope * sqft + intercept for sqft in square_footage]
for sqft, price in zip(square_footage, predicted_prices):
    print(f"Square Footage: {sqft}, Predicted Price: {price}")

```
